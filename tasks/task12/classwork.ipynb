## **Решение задач №12**

```python
for n in range(4, 10_000):
    string = "3" + "1" * n

    while "31" in string or "211" in string or "1111" in string:
        if "31" in string:
            string = string.replace("31", "1", 1)
        if "211" in string:
            string = string.replace("211", "13", 1)
        if "1111" in string:
            string = string.replace("1111", "2", 1)

    if sum(map(int, string)) == 15:
        print(n)
        break

# Answer: 50
```

Как видно из условия, знак ">" (больше) выступает в роли своеобразного "курсора", перескакивая через 1, получается **две двойки**, через 2 - **одну двойку**, через 0 - **одну единицу**.

Приведем следующую таблицу

|     | $"0" \times 37$ | $"1" \times n$ | $"2" \times 37$ |
| :---: | :---: | :---: | :---: |
| $"0"$ |  $0$  |  $0$  |  $0$  | 
| $"1"$ |  $1$  |  $0$  |  $0$  |
| $"2"$ |  $0$  |  $2$  |  $1$  |

Тогда итоговая сумма цифра равна $37 \times 1 \times 1 + n \times 2 \times 2 + 37 \times 1 \times 2 = 37 \times 3 + n \times 4 = 111 + n \times 4$

Простые числа, большие $111$: $113, 127, 131, 137, ...$\
Так как $n \neq 0$, переберем значения для $n \gt 0$: $115, 119, 123, 127, ...$\
Видим, что число $127$ - первое совпадение, $n$ при котором это достигается - это число $3$

**Ответ: 3**
