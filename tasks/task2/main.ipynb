# **Решение задач №2 из ЕГЭ**

## **№1**

```python
print("x y z w")
for x in 0, 1:
    for y in 0, 1:
        for z in 0, 1:
            for w in 0, 1:
                if (not ((not x or y) and not w) or not (z and not(y and w))) == 0:
                    print(x, y, z, w)
# Answer: yxwz
```

### **Решение с помощью `itertools.product`**

```python
from itertools import product

print("x y z w")
for x, y, z, w in product((0, 1), repeat=4):
    if (not ((not x or y) and not w) or not (z and not (y and w))) == 0:
        print(x, y, z, w)
# Answer: yxwz
```

## **№2**

```python
print("x y z w")

for x in 0, 1:
    for y in 0, 1:
        for z in 0, 1:
            for w in 0, 1:
                if not(w <= (x == (y or y))) and (z <= x):
                    print(x, y, z, w)
# Answer: yxwz
```
