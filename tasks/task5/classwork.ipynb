```python
answers = []
for n in range(10, 10_000):
    input_num = n

    # Перевод из 10СС в 3СС
    raw_n = ""
    while n != 0:
        raw_n += str(n % 3)
        n //= 3
    raw_n = raw_n[::-1]

    if input_num % 4 == 0:
        raw_n += raw_n[-3:]
    else:
        raw_n = "1" + raw_n + "20"

    R = int(raw_n, 3)
    if R > 423:
        answers.append(R)

print(min(answers))
# Answer: 438
```

```python
def build_num(n):
    r_num = bin(n)[2:]
    if n % 3 == 0:
        r_num += r_num[-2:]
    else:
        r_num = "1" + r_num + "1"

    return int(r_num, 2)


answers = []
for n in range(1, 10_000):
    R = build_num(n)
    if R > 700:
        answers.append(R)

print(min(answers))
# Answer: 709
```

```python
def build_num(num):
    r = bin(num)[2:]
    r = r.replace("1", "*")
    r = r.replace("0", "01")
    r = r.replace("*", "10")
    
    return int(r, 2)


answers = []
for n in range(1, 10_000):
    R = build_num(n)
    if R < 256:
        answers.append(R)

print(max(answers))
# Answer: 170
```

```python
from itertools import permutations


def is_prime(num):
    div = 2
    while div < num:
        if num % div == 0:
            return False
        div += 1
    return True


def build_num(num):
    num = str(num)

    two_digit_numbers = [
        x for x in permutations(num, 2)
        if x[0] != "0" and is_prime(int("".join(x)))
    ]
    
    return len(two_digit_numbers)


max_amount = 0
answer = None
for n in range(100, 1000):
    R = build_num(n)
    if R >= max_amount:
        max_amount = R
        answer = n

print(answer)
# Answer: 731
```
