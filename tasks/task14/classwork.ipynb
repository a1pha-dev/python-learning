```python
from string import ascii_uppercase


for x in "0123456789" + ascii_uppercase[:11]:
    value = int(f"82934{x}2", 21) + int(f"2924{x}{x}7", 21) + int(f"67564{x}8", 21)

    if not value % 20:
        print(value // 20)
        break

# Answer: 72450445
```

```python
def to_five_numeric(num):
    answer = ""
    while num:
        answer += str(num % 5)
        num //= 5
    return answer[::-1]


amount_of_zeros = {}
for x in range(1, 2006):
    amount_of_zeros[to_five_numeric(x).count("0")] = x

print(amount_of_zeros[max(amount_of_zeros)])
# Answer: 1875
```

```python
values = {
    x: {"M": int(f"432{x}3", 15), "N": int(f"86{x}86", 15)} for x in "0123456789ABCDE"
}

a = 1
end = False
while not end:
    for value in values.values():
        if not (value["M"] + a) % value["N"]:
            print(a)
            end = True
            break
    a += 1

# Answer: 212298
```
