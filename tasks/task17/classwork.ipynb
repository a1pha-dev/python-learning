## **Классная работа**

```python
with open("data/17_21712.txt") as file:
    seq = list(map(int, file))

sub_min = min(num for num in seq if 1000 <= num < 10_000 and num % 10 == 6)

triples = [
    sum(seq[i : i + 3])
    for i in range(len(seq) - 2)
    if sum(1000 <= abs(num) < 10_000 and abs(num) % 10 == 6 for num in seq[i : i + 3]) == 1
    and sum(seq[i : i + 3]) <= sub_min
]
print(len(triples), max(triples))
# Answer: 507 1042
```
