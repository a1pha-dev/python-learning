## **Решение задач №16**

```python
from sys import setrecursionlimit


setrecursionlimit(7000)


def f(n):
    if n < 20:
        return n
    return (n - 6) * f(n - 7)


print((f(47_872) - 290 * f(47_865)) / f(47_858))
# Answer: 2276939784
```

```python
f = {n: n for n in range(20)}

for n in range(20, 47_873):
    f[n] = (n - 6) * f[n - 7]

print((f[47_872] - 290 * f[47_865]) / f[47_858])
# 2276939784
```

Приведем аналитическое решение:

$\frac{f(47872) - 290 \times f(47865)}{f(47858)}=\frac{(47872 - 6) \times f(47865) - 290 \times f(47865)}{f(47858)}=\frac{47576 \times f(47865)}{f(47858)}=\frac{47576 \times 47859 \times f(47858)}{f(47858)}=$

$=47576 \times 47859 = 2276939784$

**Ответ: 2276939784**
