## **Решение задач №9 из ЕГЭ**

```python
with open("files/9_21592.txt") as file:
    table = [list(map(int, row.split())) for row in file]

for i, row in enumerate(table):
    repeated = [num for num in row if row.count(num) == 2]
    unrepeated = [num for num in row if row.count(num) == 1]
    if len(repeated) == 6 and len(unrepeated) == 2 and (max(repeated) - min(repeated)) ** 2 > 2 * (unrepeated[0]**2 + unrepeated[1]**2):
        max_number = i + 1

print(max_number)
# Answer: 29938
```

```python
with open("files/9_19241.txt") as file:
    table = [list(map(int, row.split())) for row in file]

for i, row in enumerate(table):
    repeated = [num for num in row if row.count(num) == 3]
    unrepeated = [num for num in row if row.count(num) == 1]
    if len(repeated) == 6 and len(unrepeated) == 1 and sum(repeated) / 6 < unrepeated[0]:
        max_number = i + 1

print(max_number)
# Answer: 17975
```

```python
with open("files/9_14253.txt") as file:
    table = [list(map(int, row.split())) for row in file]

amount = 0
for row in table:
    repeated = [num for num in row if row.count(num) == 2]
    unrepeated = [num for num in row if row.count(num) == 1]
    if len(repeated) == 6 and len(unrepeated) == 1:
        amount += 1
        continue
    average = sum(row) / 7
    if int(average**0.5) ** 2 == average:
        amount += 1

print(amount)
# Answer: 6825
```
