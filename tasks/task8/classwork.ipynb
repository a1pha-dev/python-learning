```python
from itertools import product

number = 0
for i, word in enumerate(product(sorted("АРГУМЕНТ"), repeat=4)):
    # В данном условии важно привести элемент итератора product - word - к списку, 
    # тогда оператор сравнения сработает корректно 
    if len(set(word)) == len(word) and sorted(word) == list(word):
        number = i + 1

print(number)
# Answer: 2424
```

```python
from itertools import product

odd_digits = "1357"
even_digits = "2468"

odd_even = [odd_digits, even_digits] * 4 + [odd_digits]
even_odd = [even_digits, odd_digits] * 4 + [even_digits]

amount = 0

for num in tuple(product(*odd_even)) + tuple(product(*even_odd)):
    if all(num.count(char) <= 3 for char in num):
        amount += 1

print(amount)
# Answer: 483840
```
