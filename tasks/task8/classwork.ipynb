# **Решение задач №8**

```python
from itertools import product


for i, word in enumerate(sorted(product("ГИРЛЯНДА", repeat=6))):
    if i % 2 and word[0] != "Я" and word.count("Д") == 3:
        answer = i

print(answer)
# Answer: 226456
```

### **Приведем другое решение**

Сопоставим буквам *Г*, *И*, *Р*, *Л*, *Я*, *Н*, *Д*, *А* цифры 8-ричной СС, упорядочив их в алфавитном порядке:
- А - 0
- Г - 1
- Д - 2
- И - 3
- Л - 4
- Н - 5
- Р - 6
- Я - 7

Выходит, что искомое слово соответствует наибольшему нечетному 8-ричному шестизначному числу, не начинающееся с цифры 7 и содержащее ровно три цифры 2, т.е. $672227$, переведем в 10-тичную СС, получим:

```python
int("672227", 8)
```

Поскольку число меньше его номера на 1, то ответ - 226456

**Ответ: 226456**

```python
from itertools import product


amount = 0
for num in product("0123456789ABCDE", repeat=5):
    odd_digits = num[::2]
    even_digits = num[1::2]
    if num[0] != "0" and (
        all(not int(digit, 15) % 2 for digit in odd_digits)
        and all(not int(digit, 15) % 3 for digit in even_digits)
        or all(not int(digit, 15) % 3 for digit in odd_digits)
        and all(not int(digit, 15) % 2 for digit in even_digits)
    ):
        amount += 1

print(amount)
# Answer: 17438
```

```python
from itertools import product


amount = 0
for word in product("МАСЛО", repeat=6):
    if word.count("А") + word.count("О") == 1:
        amount += 1

print(amount)
# Answer: 2916
```

### **Приведем другое решение**
Поскольку гласная буква может быть одна, то возможны следующие варианты:
- *ГССССС*
- *СГССС*
- *ССГССС*
- ...
- *СССССГ*

(*С* - согласная буква, *Г* - гласная)\
Итого $6$ вариантов. Поскольку гласных $2$ штуки, то всего вариантов размещений гласных $6 \times 2 = 12$ штук.\
На месте каждой из $5$ согласных букв может стоять $3$ буквы: *М*, *С*, *Л*, поэтому согласно правилу комбинаторного умножения получаем: $12 \times 3^5 = 2916$\
**Ответ: 2916**
