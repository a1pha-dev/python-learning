## **Решение задач №15**

```python
def f(x, a):
    return not x % 7 or x % 13 or x > a - 40


for a in range(1000, 0, -1):
    if all(f(x, a) for x in range(1, 1000)):
        print(a)
        break
# Answer: 52
```

```python
def f(x, y, a):
    return 5 < y or x > 32 or x + 2 * y < a


for a in range(1000):
    if all(f(x, y, a) for x in range(1000) for y in range(1000)):
        print(a)
        break
# Answer: 43
```

```python
def f(x, a):
    return (x % 10 != 5 and x % 10 == 4) <= (x > a - 11)


for a in range(1000, 0, -1):
    if all(f(x, a) for x in range(1, 1000)):
        print(a)
        break
# Answer: 14
```
